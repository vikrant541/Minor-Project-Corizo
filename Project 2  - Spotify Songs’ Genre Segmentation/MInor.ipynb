```python
# Import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
most_popular = df_tracks.query('track_popularity> 90', inplace  = False).sort_values('track_popularity', ascending = False)
most_popular[:10]
```

```python
df_tracks.set_index("track_album_release_date", inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks['duration'] = df_tracks["duration_ms"].apply( lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
df_tracks.duration.head()
```

```python
corr_df=df_tracks.drop(["key","mode" ], axis=1).corr(method="pearson")

plt.figure(figsize=(14,6))

heatmap=sns.heatmap(corr_df,annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths = 1, linecolor = "Black")

heatmap.set_title("Correlation HeatMap Between Variable")

heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90) 
```

```python
import pandas as pd
from sklearn.cluster import KMeans

# Replace "My Drive/Spotify Dataset.csv" with your actual path
df = pd.read_csv("spotify dataset.csv")

# Choose features for clustering (replace with your desired features)
features = ["speechiness","tempo","instrumentalness","danceability"]
```

```python
#To check data types of each column and perhaps null values
df_tracks.info()
```

```python
# Ensure features are numerical (convert if needed)
df_numerical = df[features]
```

```python
# Use the standardized data if chosen above, otherwise use df_numerical
kmeans = KMeans(n_clusters=3)
kmeans.fit(df_numerical)  # Replace with df_numerical if no standardization

cluster_labels = kmeans.labels_
df["cluster"] = cluster_labels
```

```python
# Visualize clusters with features like danceability and energy
sns.scatterplot(
    x="energy",
    y="loudness",
    hue="cluster",
    data=df,
)
plt.show()

# Explore other features and compare across clusters
df.groupby("cluster").mean()

# Consider dimensionality reduction (PCA) for better visualization if needed
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(df_numerical, cluster_labels)
print("Silhouette Coefficient:", silhouette_avg)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(df_numerical, cluster_labels)

# Predict cluster for a new song
new_song_features = [00.120489,145.875853,0.069922,0.665404]  # Replace with new song's features
predicted_cluster = clf.predict([new_song_features])[0]
print("Predicted Cluster:", predicted_cluster)
```
